### Write a program to display sequence fibonacci using Recursion

```python
n = int(input())
fibo(n)
```

```python
def fibo(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  return n + fibo(n)
```

### Write a program to print factorial using Recursion

```python
def fact(n):
  if n == 0 or n == 1:
    return 1
  return n * fact(n-1)
```

```python
n = int(input())
print(fact(n))
```

### Write a python program for cube sum of first n natural numbers?

```python
def cubeSum(a):
  sum = 0
  for i in range(2,a+1):
    sum = sum + i**3
  return sum
```

```python
a = int(input("Enter the number to print cube sum"))
print(cubeSum(a))
```
