### Write a program to check if the given number is disarium number?

```python
def get_digits(num):
  digits = []
  while num != 0:
    rem = num % 10
    num = int(num /10)
    digits.append(rem)
  return digits[::-1]

def check_disarium(num):
  if num < 0 or type(num) != int:
    raise Exception("Plese enter a positive integer number")
  digits = get_digits(num)
  sum = 0
  for i in range(1, len(digits) + 1):
    sum = sum + (digits[i-1] ** i)
  if sum == num:
    return True
  return False

test = [135, 89, 80, 175]
output = []
try:
  for i in test:
    output.append(check_disarium(i))
  for i in output:
    if i:
      print("Is a disarium number")
    else:
      print("not a disarium number")
except Exception as e:
  print(e)
```

### Write a program to find all disarium numbers between 1 and 100

```python
try:
  disarium_1_100 = filter(check_disarium, [i for i in range(1, 101)])
  print("The disarium numbers between 1-100 are: ")
  for i in disarium_1_100:
    print(i)
except Exception as e:
  print(e)
```

### Write a python program to check if the given number is Happy number

```python
def get_digits(num):
  digits = []
  while num != 0:
    rem = num % 10
    num = int(num /10)
    digits.append(rem)
  return digits[::-1]

def to_square(a):
  return a**2

def check(l, num):
  if sum(l) == num:
    return sum(l), 1
  if sum(l) == 1:
    return sum(l), 2
  return sum(l), 3

def check_happy(num1, num, flag = None):
  l = list(map(to_square, get_digits(num1)))
  num1, check_val = check(l, num)
  if check_val == 1 and flag == None:
    return 1
  elif check_val == 2:
    return 2
  else:
    return check_happy(num1, num)
def final_check(result):
  if result == 1:
    print("Its a unhappy number")
  elif result == 2:
    print("Its a happy number")
  else:
    print("Not a happy number neither a unhappy one")
```

```python
result = 0
try:
  result = check_happy(4, 3, 1)
except:
  result = 3
final_check(result)
```

### Write a python program  to print all happy numbers between 1 and 100

```python
happy_num = [1]
for i in range(2, 101):
  try:
    result = check_happy(i, i, 1)
  except:
    result = 3
  if result == 2:
    happy_num.append(i)
print(happy_num)
```

### Write a python program to check if a number is harshad number

```python
def get_digits(num):
  digits = []
  while num != 0:
    rem = num % 10
    num = int(num /10)
    digits.append(rem)
  return digits[::-1]
def check_harshad(num):
  if num < 0 or type(num) != int:
    raise Exception("Plese enter a positive integer number")
  digits = get_digits(num)
  if num % sum(digits) == 0:
    print("Is a harshad number")
  else:
    print("Not a harshad number")
```

```python
num = 171
try:
  check_harshad(num)
except Exception as e:
  print(e)
```

### Write a program to print all pronic numbers between 1 and 100

```python
def check_pronic(num):
  i = 1
  while (i * (i + 1)) <= num:
    if (i * (i + 1)) == num:
      return True
    i = i + 1
  return False
list_pronic_1_100 = list(filter(check_pronic, [num for num in range(1, 101)]))
print(list_pronic_1_100)
```
