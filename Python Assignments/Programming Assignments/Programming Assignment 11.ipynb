### Write a python Program to find words which are greate than given length k

```python
def rem_punctuations(string_1):
  punctuation_list = [',', ';',':',"'", '"', '?', '-', '!', '.', '(', ')', '{', '}', '[', ']']
  string_2 = ""
  for i in string_1:
    if i not in punctuation_list:
      string_2 += i
  return string_2

def find_words(string_1, k):
  if k>len(string_1):
    raise Exception("K can't be more than length of the string")
  string_1 = rem_punctuations(string_1)
  final_word_list = []
  for i in string_1.split():
    if len(i) > k:
      final_word_list.append(i)
  return final_word_list

k = 3
string_1 = "hello there this is me, trying out this question."
try:
  print(find_words(string_1, k))
except Exception as e:
  print(e)
```

### Write a python program to check if a given string is binary string or not

```python
def check_binary_string(string_1):
  check_list = ['1', '0', ' ']
  flag = True
  for i in string_1:
    if i not in check_list:
      flag = False
      break
  return flag

string_1 = '1100111 1100101 1100101 1101011 1110011'
print("It is a binary string" if check_binary_string(string_1) else "It is not a binary string")
```

### Write a python program to find all duplicate characters in string

```python
def rem_punctuations(string_1):
  punctuation_list = [',', ';',':',"'", '"', '?', '-', '!', '.', '(', ')', '{', '}', '[', ']', ' ']
  string_2 = ""
  for i in string_1:
    if i not in punctuation_list:
      string_2 += i
  return string_2
def find_duplicate(string_1):
  string_1 = rem_punctuations(string_1)
  dict_count = {}
  duplicate_char_list = []
  for i in string_1:
    if i not in dict_count.keys():
      dict_count[i] = 1
    else:
      dict_count[i] += 1
  for k, v in dict_count.items():
    if dict_count[k] > 1:
      duplicate_char_list.append(k)
  return duplicate_char_list

string_1 = "hello, there this, is me, trying out, this question."
print("The duplicate characters are: ", find_duplicate(string_1))
```

### Write a python program to check if a string contains any special character

```python
def check_symbol(string_1):
  if len(string_1) == 0:
    raise Exception("String Lenght cannot be zero")
  symbol_string = "!#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~"
  symbol_list = ['"'] + [i for i in symbol_string]
  for i in string_1:
    if i in symbol_list:
      return True
  return False

string_1 = "hello there this is me trying out this question"
string_2 = "hello, there this, is me, trying out, this question."
try:
  print("Contains symbol" if check_symbol(string_1) else "Doesn't contain symbol")
  print("Contains symbol" if check_symbol(string_2) else "Doesn't contain symbol")
except Exception as e:
  print(e)
```
