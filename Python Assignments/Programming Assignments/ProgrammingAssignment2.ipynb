Write a python program to solve quadratic equation?

```python
var_a = int(input())
var_b = int(input())
var_c = int(input())
discriminant = (var_b ** 2) - (4*var_a*var_c)
if discriminant > 0:
  root_1 = ((-var_b) + (discriminant ** 0.5)) / (2*var_a)
  root_2 = ((-var_b) - (discriminant ** 0.5)) / (2*var_a)
elif discriminant == 0:
  root_1 = root_2 = ((-var_b) + (discriminant ** 0.5)) / (2*var_a)
else:
  print("the roots are unqual and imaginary")
print("Roots are {} and {}".format(root_1, root_2))
```
