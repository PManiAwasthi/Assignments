```python
def find_len(arr):
  length = []
  length.append(len(arr))
  for i in (0, len(arr)-1):
    if type(arr[i]) == list:
      length.append(find_len(arr[i]))
  return length

def flattenList(nestedList):
  if not(bool(nestedList)):
    return nestedList
  if isinstance(nestedList[0], list):
    return flattenList(*nestedList[:1]) + flattenList(nestedList[1:])
  return nestedList[:1] + flattenList(nestedList[1:])
```

```python
def check_length(arr1, arr2):
  length1 = flattenList(find_len(arr1))
  length2 = flattenList(find_len(arr2))
  if length1 == length2 and length1[0] == 2 and length2[0] == 2:
    return True
  else:
    return False
```

```python
def print_ar(arr):
  for i in range(len(arr)):
    for j in range(len(arr[0])):
      print(arr[i][j], end = " ")
    print()
```

### Write a python program to multiply 2 matrices

```python
a = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]
# 3x4 matrix
c = [[5,8,1,2],
    [6,7,3,0],
    [4,5,9,1]]
try:
  if len(a[0]) == len(c):
    mul_arr = [[0 for i in range(len(c[0]))] for j in range(len(a))]
    for i in range(len(a)):
      for j in range(len(c[0])):
        for k in range(len(c)):
          mul_arr[i][j] += a[i][k] * c[k][j]
    print_ar(mul_arr)
  else:
    raise Exception("Column of first matrix should be equal to rows of the other")
except Exception as e:
  print(e)
```

### Write a python program to transpose a matrix

```python
a = [[12,7],
    [4 ,5],
    [3 ,8]]
try:
  b = [[0 for j in range(len(a))] for i in range(len(a[0]))]
  for i in range(len(a)):
    for j in range(len(a[0])):
      b[j][i] = a[i][j]
  for i in b:
    print(i)
except Exception as e:
  print(e)
```

### Write a program to sort words in alphabetic order

```python
string_1 = 'The criminal, said the judge, was an idiot.'
words = [i.lower() for i in string_1.split()]
words.sort()
print(words)
```

### Write a python program to remove punctuation from  a string

```python
list_punct = ['.', ',', ';', "'", '"', "?", '!', '(', ')', '-', '_', '~', ':', '[', ']', '{', '}']
string_1 = 'The criminal, said the judge, was an idiot.'
string_2 = ''
for i in string_1:
  if i not in list_punct:
    string_2 += i
print(string_2)
```
