### Write a python program to extract unique values from dictionary values

```python
dict_1 = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 1: 'apple', 2: 'ball',3: (4,3,2)}
unique_values = []
for i in dict_1.values():
  if i not in unique_values:
    unique_values.append(i)
unique_values = [i for i in set(unique_values)]
print(unique_values) ## value which are list, set, etc are treated as 1 single element
```

### Write a python program to merge 2 dictionaries

```python
def merging_dict(dict_1, keys_list, dict_merged):
  if len(dict_1.keys()) == 0:
    return dict_merged
  for k in keys_list:
    if k in dict_1.keys():
      if k not in dict_merged.keys():
        dict_merged[k] = dict_1[k]
      else:
        if type(dict_merged[k])!=list:
          dict_merged[k] = [dict_merged[k]]
        if hasattr(dict_1[k], '__iter__'):
          if type(dict_1[k]) == dict:
            for i in dict_1[k].values():
              dict_merged[k].append(i)
          elif type(dict_1[k]) == list:
            for i in dict_1[k]:
              dict_merged[k].append(i)
          else:
            dict_merged[k].append(dict_1[k])
        else:
          dict_merged[k].append(dict_1[k])
  return dict_merged

def merged_dict(dict_1, dict_2):
  keys_list = list(set(list(dict_1.keys()) + list(dict_2.keys())))
  dict_merged = merging_dict(dict_1, keys_list, {})
  dict_merged = merging_dict(dict_2, keys_list, dict_merged)
  return dict_merged

dict_1 = {1: 'apple', 2: 'ball',3: (4,3,2)}
dict_2 = {'name': 'John', 1: [2, 4, 3], 3: (4,3,2)}
print(merged_dict(dict_1, dict_2))
#this code can also work if the dictionary contains list, dictionary or set in the values
```

### Write a Python program to convert key-values list to flat dictionary?

```python
key_val_list = [
    {'key1': 1},
    {'key2': 2},
    {'key3': 3},
    {'key4': 4},
    {'key5': 5},
    {'key2': 6},
    {'key3': 7} 
]
```

```python
dict_1 = {}
for i in key_val_list:
  if type(i) == dict:
    for k in i.keys():
      if k not in dict_1.keys():
        dict_1[k] = i[k]
      else:
        if type(dict_1[k]) != list:
          dict_1[k] = [dict_1[k]]
        dict_1[k].append(i[k])
print(dict_1)
        
```
