### Write a python program to find sum of array

```python
arr = input('Enter the array elements: ').split()
arr = [int(x) for x in arr]
sum = 0
for i in arr:
  sum = sum + i
print(sum)
```

### Write a python program to find largest element in an array

```python
arr = input('Enter the array elements: ').split()
arr = [int(x) for x in arr]
max = arr[0]
for i in arr[1:]:
  if i > max:
    max = i
print("Max = ", max)
```

### Write a python program for array rotation

```python
arr = input('Enter the array elements: ').split()
arr = [int(x) for x in arr]
```

```python
def left_rotate(rotation, arr):
  if(rotation > 0):
    raise Exception("for left rotation the value of rotation should be negative")
  rotation = abs(rotation)
  for i in range(0,rotation):
    arr = rotate_L(arr)
  return arr

def rotate_L(arr):
  temp = arr[0]
  temp_ar = [0] * len(arr)
  for i in range(0, len(arr) - 1):
    temp_ar[i] = arr[i+1]
  temp_ar[-1] = temp
  return temp_ar

def right_rotate(rotation, arr):
  if(rotation < 0):
    raise Exception("for rigth rotation the value of rotation should be positive")
  for i in range(0, rotation):
    arr = rotate_R(arr)
  return arr

def rotate_R(arr):
  temp = arr[-1]
  temp_ar = [0] * len(arr)
  for i in range(0, len(arr) - 1):
    temp_ar[i + 1] = arr[i]
  temp_ar[0] = temp
  return temp_ar
```

```python
# arr = [1, 2, 3, 4, 5, 6]
type_of_rotation = int(input("1.Left rotation\n2.Right rotation\n"))
while True:
  try:
    if type_of_rotation == 1:
      rotation = int(input("Enter rotation value: "))
      arr = left_rotate(rotation, arr)
      break
    elif type_of_rotation == 2:
      rotation = int(input("Enter the rotation value: "))
      arr = right_rotate(rotation, arr)
      break
  except Exception as e:
    print(e)
print(arr)
```
