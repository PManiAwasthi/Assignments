### Write a python program to find sum of elements in list

```python
list_a = [int(i) for i in input().split()]
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  sum = 0
  for i in list_a:
    sum += i
  print(sum)
except Exception as e:
  print(e)
```

### Write a python program to multiply all numbers in the list

```python
list_a = [int(i) for i in input().split()]
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  mul = 1
  for i in list_a:
    mul *= i
  print(mul)
except Exception as e:
  print(e)
```

### Write a python program to find the smallest number in a list

```python
list_a = [int(i) for i in input().split()]
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  min = list_a[0]
  for i in list_a:
    if min > i:
      min = i
  print(min)
except Exception as e:
  print(e)
```

### Write a program to find the largest number in a list

```python
list_a = [int(i) for i in input().split()]
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  max = list_a[0]
  for i in list_a[1:]:
    if max < i:
      max = i
  print(max)
except Exception as e:
  print(e)
```

### Write a python program to find Nth largest number in a list

```python
list_a = [23,41,232,31,28]
def get_max(list_a, count):
  if count - 1 == 0:
    return find_max(list_a)
  return get_max(remove_max(list_a, find_max(list_a)), count -1)
def find_max(list_a):
  max = list_a[0]
  for i in list_a:
    if max < i:
      max = i
  return max
def remove_max(list_a, max):
  list_b = [i for i in list_a if i!= max]
  return list_b

list_a = [23,41,232,31,28]
nth_max = int(input('Enter the which nth max element is required: '))
if nth_max != 0 and nth_max < len(list_a):
  print(get_max(list_a, nth_max))
else:
  print("Wrong input")
```

### Write a python program to print even numbers in a list

```python
list_a = [23,41,232,31,28]
def is_even(num):
  if num == 0:
    return False
  if num % 2 == 0:
    return True
  return False
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  even_list = list(filter(is_even, list_a))
  print(even_list)
except Exception as e:
  print(e)
```

### Write a python program to print odd numbers in a list

```python
list_a = [23,41,232,31,28]
def is_odd(num):
  if num == 0:
    return False
  if num % 2 == 0:
    return False
  return True
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  odd_list = list(filter(is_odd, list_a))
  print(odd_list)
except Exception as e:
  print(e)
```

### Write a program to remove empty list from list

```python
list_a = [23,41,232,31,[], []]
list_without_empty_list = []
try:
  if len(list_a) == 0:
    raise Exception("List can't be empty")
  for i in range(0, len(list_a)):
    if type(list_a[i]) == list and len(list_a[i]) == 0:
      continue
    list_without_empty_list.append(list_a[i])
  print(list_without_empty_list)
except Exception as e:
  print(e)
```

### Write a python program to count occurences of an element in a list

```python
dict_count = {}
list_a = [23,41,23,31,28, 'hello', 'hello']
dict_count = {str(k):v for k,v in zip(list_a, [0] * len(list_a))}
for i in list_a:
  dict_count[str(i)] += 1
for k, v in dict_count.items():
  print(k, v)
```
