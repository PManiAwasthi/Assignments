Write a program to find the factorial of a number?

```python
var_num = int(input())
factorial = 1
for i in range(2, var_num + 1):
  factorial = factorial * i
print("Factorial of {} is {}".format(var_num, factorial))
```

Write a progrma to display the multiplication table?

```python
gen_table_of = int(input())
for i in range(1, 11):
  print("{} x {} = {}".format(gen_table_of, i, gen_table_of * i))
```

Write a program to print fabonacci sequence?

```python
gen_fab_of = int(input())
num_a = 0
num_b = 1
for i in range(1, gen_fab_of+1):
  print(num_a)
  sum = num_a + num_b
  num_a = num_b
  num_b = sum
```

Write a program to check armstrong number?




```python
num_to_check = int(input())
sum = 0
iter = num_to_check
while iter != 0:
  rem = iter % 10
  sum = sum + (rem ** 3)
  iter = int(iter / 10)
if sum == num_to_check:
  print("Is armstrong")
```

Write a program to find the sum of natural numbers?

```python
sum_upto = int(input())
sum = 0
for i in range(1, sum_upto + 1):
  sum = sum + i
print("sum of natural numbers upto {} is {}".format(sum_upto, sum))
```
