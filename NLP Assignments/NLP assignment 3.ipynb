Explain the basic architecture of RNN cell.

```python
"""
It is a type of neural network that is designed to process sequential data commonly used in NLP and other such tasks that involve analyzing sequential data.
Basic architecture include:
- Input layer: takes in input usually in form of sequence of data points
- Hidden state: A representation of previous input data that is carried over from one time step to the next. Used to maintain contexxt and retain information about the previous 
inputs as RNN processes the input sequence.
- activation function: used to introduce non lineariry to the RNN and allows it to learn complex pattern in the data.
- Output layer: gives the final output of the RNN based on the inputp data and hidden state.
- recurrent connections: it allows the hidden state to be carried over from one time step to the next.
"""
```

Explain backpropogation through time (BPTT)

```python
"""
It is an algorithm which is used to train the recurrent neural networks, which are a type of NNs able to work with sequential data. It is a variant of backpropogation algorithm 
which is widely used method of traning a feed forward NN.
It calculates the gradient of the loss function with respect to the weights of the RNN at each time step in the input sequence and these loss functions measure the difference 
between the predicted output and the true oputut and are later minimized.
"""
```

Explain vanishing and exploding gradient?

```python
"""
The vanishing and exploding gradient are the problems which can occurs when training a DNN, including RNN. It is a type of problem that occurs when the gradient of the loss 
function with respect to the weights is too small or large respectively. When the gradient is too small its called vanishing gradient and if its too large it is exploding gradient 
and if vanishing gradient occurs the learning stops as the weights and other parameters just either stop updating or very slow or negligible change is observed which is highly 
undesirable, similarly is the exploding gradient occurs the updates will be too drastic and more or less makes the model unstabel and oscillate widely.
"""
```

Explain long short term memory?

```python
"""
A type of RNN that is designed to process sequential data and maintain context over a longer period of time. It consists of input gate, forgot gate, output gate and a memory cell.
These all work together to control the flow of information into and out of the cell, and make the decision of retaining or releasing the information stored in the cell.

The input gate, forgot gate, and the ouput gate are all sigmoid functions that take the input data and hidden state and produce output between 0 and 1. The memory cell is a 
linear function that takes in the inpupt and hidden state and procuded the output. The input gate controls what information is allowed to enter the memory cell and the 
forgot gate controls the information from the previous hidden state is retained or not, the output gate controls the info from the memory cell to be allowed to pass through 
as the output of the LSTM.
"""
```

Explain gated recurrent unit?

```python
"""
It is similar to the LSTM as it is a type of RNN that is able to process seqentail data and maintain context over a longer period of time, but has a simpler architecture than 
LSTM which makes it typically easier to train.

The architecture consists of 2 interacting components: the reset gate and an update gate. the reset gate and update gates are both sigmoid function that takes in the input data 
and hidden state and produce output values ranging 0 and 1.
The  reset gate controls if the information of the hidden state is to be retained or not, and the update gate controls what information from input data is allowed to enter the 
current hidden state. The current hidden state is calculated as a combination of the previous hidden state and input data, with the influence of each controlled by the reset and 
update gates.
"""
```
